# **Реализация подсчета дистанции Левенштейна**

Необходимо реализовать подсчет дистанции Левенштейна.

Для удобства можно весь алгоритм записать в виде функции, принимающий набор аргументов:

* Первая последовательность нуклеотидов
* Вторая последовательность нуклеотидов

Функция должна возвращать дистанцию между последовательностями в численном виде.

```
# Заданные последовательности:

seq1 = TGG
seq2 = TAGC

# Возвращаемая дистанция:
distance = 2
```

````python
seq_1 = 'TCATAG'
seq_2 = 'TCAG'
````

````python
def dist_levenshtein(seq1, seq2):    
    # Инициализация матрицы
    m, n = len(seq1), len(seq2)
    score_matrix = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    
    # Заполнение первой строки и первого столбца
    for i in range(m + 1):
        score_matrix[i][0] = i
    for j in range(n + 1):
        score_matrix[0][j] = j
    
    # Заполнение матрицы
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Если символы совпадают, стоимость = 0, иначе 1 (замена)
            match = score_matrix[i-1][j-1] + (0 if seq1[i-1] == seq2[j-1] else 1)
            delete = score_matrix[i-1][j] + 1
            insert = score_matrix[i][j-1] + 1
            
            # Выбор минимального значения
            score_matrix[i][j] = min(match, delete, insert)
       
    return score_matrix[m][n]
````

````python
dist_levenshtein(seq_1, seq_2)
````
